## Quick plot

A simple sine wave to confirm plotting works inline.

```python
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

plt.figure(figsize=(6, 3))
plt.plot(x, y)
plt.title("sin(x)")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.tight_layout()
plt.show()
```
